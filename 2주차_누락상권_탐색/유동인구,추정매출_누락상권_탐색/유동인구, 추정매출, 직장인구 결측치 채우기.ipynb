# 유동인구, 직장인구, 추정매출 결측치 채우기_문지용

```python
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
data_path = '../../직장인구_전체통합+/'
df = pd.read_csv(data_path + 'data_processed_final.csv')
```

```python
# 유동인구 데이터
df_people = df[['기준_년_코드', '기준_분기_코드', '상권_코드', '총_유동인구_수']]
```

```python
# 직장인구 데이터
df_job = df[['기준_년_코드', '기준_분기_코드', '상권_코드', '총_직장_인구_수']]
```

```python
# 추정매출 데이터
df_sales = df[['기준_년_코드', '기준_분기_코드', '상권_코드', '당월_매출_금액']]
```

### 유동인구 결측치 채우기

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder()
x = onehot.fit_transform(df_people['상권_코드'].values.reshape(-1,1)).toarray()
```

```python
onehot_df = pd.DataFrame(x, columns = ["상권코드_"+str(int(i)) for i in range(x.shape[1])])
```

```python
people_o_hot = pd.concat([df_people, onehot_df], axis=1)
```

```python
people_o_hot
```

```python
people_name_list = list(people_o_hot['상권_코드'])
people_o_hot = people_o_hot.drop(['상권_코드'], axis=1)
```

```python
import time
start = time.time()  # 시작 시간 저장
 
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors = 150)

people_o_hot_knn = imputer.fit_transform(people_o_hot)

print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```

```python
df_people_after = pd.DataFrame(data = people_o_hot_knn)
```

```python
df_people_after['상권_코드'] = people_name_list
```

```python
df_people_after
```

```python
df_people_after = df_people_after[[0,1,2,'상권_코드']]
df_people_after.columns = ['기준_년_코드', '기준_분기_코드', '총_유동인구_수', '상권_코드']
```

```python
df_people_after
```

```python
df_people_after.isnull().sum()
```
