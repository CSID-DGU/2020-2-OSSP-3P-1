```python
x = [6.3,8.4,8.6,8.9,9.1,9.4,8.9,8.4,8.1,7.9,7.7,7.4,6.7,8.7,7.8]
len(x)
```

```python
y = [14.2,12,10.2,8.5,3.2,3.2,1.8,2.6,2.6,2.6,4.4,4.7,4.7,3.3,2.4]
len(y)
```

```python
import numpy as np

from sklearn.linear_model import LinearRegression
```

```python
x_in = np.array(x).reshape(-1, 1)
y_in = np.array(y).reshape(-1, 1)
```

```python
reg = LinearRegression().fit(x_in, y_in)
```

```python
from sklearn.metrics import mean_squared_error

y_true = y
y_pred = reg.predict(x_in)
y_pred
```

```python
mean_squared_error(y_true, y_pred)
```

```python
reg.coef_
```
