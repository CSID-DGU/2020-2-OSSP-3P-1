```python
def nearest(code):
    X = list(df0[df0['상권_코드'] == code]['엑스좌표_값'])[0]
    Y = list(df0[df0['상권_코드'] == code]['와이좌표_값'])[0]
    x_list = list(df0['엑스좌표_값'])
    y_list = list(df0['와이좌표_값'])
    distance_list=[]
    for i in range(1496):
        x = x_list[i]
        y = y_list[i]
        distance = ((X-x)**2) + ((Y-y)**2)
        distance_list.append(distance)
    sorted_distance_list = sorted(distance_list)
    distance = sorted_distance_list[1]
    for i in range(1496):
        if distance_list[i] == distance:
            result = i
            break
        else:
            continue
    return df0.iloc[i,3]  

def no(df,target):
    s1 = set(df0['상권_코드'])
    s2 = set(df['상권_코드'])
    sub = list(s1 - s2)
    r_list
    for i in sub:
        r = nearest(i)
        r_list.append(r)
  
```
