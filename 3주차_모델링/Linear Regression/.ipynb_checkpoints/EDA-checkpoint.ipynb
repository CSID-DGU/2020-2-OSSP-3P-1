## 1. 상관관계

### (1) 폐업률 - 추정 유동인구

```python
y = df['폐업률']
x = df['총_유동인구_수']

sns.regplot(x, y, ci = None, line_kws = {'color' : 'red'})

cor = pearsonr(x, y)

print("상관계수 : {}\np-value : {}".format(cor[0], cor[1]))
```
